```python
import pandas as pd
import numpy as np
import glob
```

```python
####    OY ###
path = r'oypath/'
all_files = glob.glob(path + "/*.csv")

dff=pd.DataFrame()

for file in all_files:
    df=pd.read_csv(file, encoding='windows-1254', skiprows=7, delimiter=';')
    df=df.dropna(how='all', axis=1)
    df=df.dropna(how='all')
    df=df.rename(columns={"Unnamed: 1": "İlçe", "Unnamed: 3": "B"})
    sehir=df.iloc[0][0]
    ilce=df.İlçe.unique()
    ilce=ilce[~pd.isnull(ilce)]
    df['Şehir']=sehir
    t=df.B.unique()
    if (t=='Belde/Köy').any()==1:
        m=df.B.isnull()==True
        df=df[m]
        dff=dff.append(df)

    elif (t=='Belde/Köy').any()==0:
        dup=['Sandık sayısı ','Kayıtlı seçmen sayısı', 'Oy kullanan seçmen sayısı', 'Geçerli oy sayısı']
        df['Şehir']=sehir
        df=df.drop_duplicates(subset=dup)
        df=df.dropna(how='all', axis=1)
        dff=dff.append(df)
dff=dff.drop(['Seçim çevresi ve ilçe', 'B'], axis=1)

##
col=dff.columns.tolist()
col=['Şehir', 'İlçe', 'Geçerli oy sayısı','Oy kullanan seçmen sayısı','Kayıtlı seçmen sayısı','Sandık sayısı ', 
     'AK PARTİ', 'MHP', 'CHP', 'İYİ PARTİ', 'HDP', 'SAADET PARTİSİ', 'CUMHUR İTTİFAKI', 
     'MİLLET İTTİFAKI', 'VATAN PARTİSİ', 'HÜDA PAR', 'BĞMZ']
fin=dff[col]
m=fin['İlçe'].str.startswith('Seçim')
dfoy=fin[~m]
dfoy=dfoy.sort_values(['Şehir', 'İlçe'])
dfoy.to_csv('2018Mv_Oy.csv', encoding='utf-8')

dfoy.head()
```
